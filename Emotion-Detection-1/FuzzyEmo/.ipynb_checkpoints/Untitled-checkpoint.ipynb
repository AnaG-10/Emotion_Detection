```python
# Run if you need to install packages (Colab)
# !pip install torch torchvision scikit-learn tqdm matplotlib

import os
import glob
from pathlib import Path
from PIL import Image
import math
import random
from tqdm import tqdm

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt

# device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
```

```python
# Update this path to your dataset root
DATA_ROOT = "YOLO_format"   # << change to your path

# helper to parse YOLO label line: class x_center y_center width height (normalized)
def yolo_to_xyxy(img_w, img_h, x, y, w, h):
    x1 = int((x - w/2) * img_w)
    y1 = int((y - h/2) * img_h)
    x2 = int((x + w/2) * img_w)
    y2 = int((y + h/2) * img_h)
    # clamp
    x1 = max(0, min(img_w-1, x1))
    y1 = max(0, min(img_h-1, y1))
    x2 = max(0, min(img_w-1, x2))
    y2 = max(0, min(img_h-1, y2))
    return x1, y1, x2, y2

def collect_samples(split):
    """
    Returns list of tuples: (image_path, (x1,y1,x2,y2), class_id)
    split in {'train','valid','test'}
    """
    samples = []
    img_dir = Path(DATA_ROOT) / split / "images"
    lbl_dir = Path(DATA_ROOT) / split / "labels"
    img_paths = sorted(img_dir.glob("*"))
    for img_path in tqdm(img_paths, desc=f"collect {split}"):
        # find corresponding label file (same stem .txt)
        lbl_path = lbl_dir / (img_path.stem + ".txt")
        try:
            img = Image.open(img_path)
            img_w, img_h = img.size
            img.close()
        except Exception as e:
            # skip unreadable
            continue

        if not lbl_path.exists():
            continue
        with open(lbl_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) < 5:
                    continue
                cls = int(float(parts[0]))
                xc, yc, bw, bh = map(float, parts[1:5])
                x1,y1,x2,y2 = yolo_to_xyxy(img_w, img_h, xc, yc, bw, bh)
                # skip degenerate boxes
                if x2 <= x1 or y2 <= y1:
                    continue
                samples.append((str(img_path), (x1,y1,x2,y2), cls))
    return samples

train_samples = collect_samples("train")
valid_samples = collect_samples("valid")
test_samples  = collect_samples("test")

print("Counts -> train:", len(train_samples), "valid:", len(valid_samples), "test:", len(test_samples))
```

```python
class FuzzyConv2D(nn.Module):
    def __init__(self, in_ch, out_ch, k=3, stride=1, padding=1):
        super().__init__()
        self.conv = nn.Conv2d(in_ch, out_ch, kernel_size=k, stride=stride, padding=padding)
        # a small per-channel trainable scale to model fuzzy membership strength
        self.alpha = nn.Parameter(torch.ones(out_ch) * 0.8)

    def forward(self, x):
        x = self.conv(x)
        # fuzzy activation: tanh(alpha * x) -- bounded, differentiable
        return torch.tanh(self.alpha.view(1, -1, 1, 1) * x)

class FuzzyEmotionNet(nn.Module):
    def __init__(self, num_classes=8):
        super().__init__()
        self.features = nn.Sequential(
            FuzzyConv2D(3, 32, k=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            FuzzyConv2D(32, 64, k=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            FuzzyConv2D(64, 128, k=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        # reduce to fixed-size vector regardless of input spatial dims
        self.pool = nn.AdaptiveAvgPool2d((4,4))
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128*4*4, 256),
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.pool(x)
        x = self.classifier(x)
        return x

# instantiate
NUM_CLASSES = 8  # change if your dataset has different classes; can infer from labels too
model = FuzzyEmotionNet(num_classes=NUM_CLASSES).to(device)
print(model)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=6, gamma=0.5)

def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    running_loss = 0.0
    total = 0
    correct = 0
    for imgs, labels in tqdm(loader, desc="train", leave=False):
        imgs = imgs.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * imgs.size(0)
        preds = outputs.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += imgs.size(0)
    return running_loss / total, correct / total

def evaluate(model, loader, criterion=None):
    model.eval()
    total = 0
    correct = 0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for imgs, labels in tqdm(loader, desc="eval", leave=False):
            imgs = imgs.to(device)
            labels = labels.to(device)
            outputs = model(imgs)
            preds = outputs.argmax(dim=1)
            correct += (preds == labels).sum().item()
            total += imgs.size(0)
            all_preds.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())
    acc = correct / total if total>0 else 0
    return acc, all_preds, all_labels
```

```python
EPOCHS = 12
best_val_acc = 0.0
ckpt_dir = Path("checkpoints")
ckpt_dir.mkdir(exist_ok=True)

for epoch in range(1, EPOCHS+1):
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion)
    val_acc, _, _ = evaluate(model, valid_loader)
    scheduler.step()

    print(f"Epoch {epoch}/{EPOCHS} - train_loss: {train_loss:.4f}, train_acc: {train_acc:.4f}, val_acc: {val_acc:.4f}")

    # save best
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), ckpt_dir / "best_fuzzy_emotion.pth")
        print("Saved best model.")
# final save
    torch.save(model.state_dict(), ckpt_dir / "last_fuzzy_emotion.pth")
print("Training finished. Best val acc:", best_val_acc)
```

```python
# load best
model.load_state_dict(torch.load(ckpt_dir / "best_fuzzy_emotion.pth"))
test_acc, preds, labels = evaluate(model, test_loader)

print("Test Accuracy:", test_acc)

# confusion matrix
cm = confusion_matrix(labels, preds)
print("Confusion matrix:\n", cm)

# classification report (precision, recall, f1)
print("Classification report:")
print(classification_report(labels, preds, digits=4))
```

```python
# Paste & run this cell in your notebook
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import json
import torch
import os

def plot_confusion_matrix(labels, preds, class_labels=None, normalize=False, figsize=(10,8), cmap="Blues"):
    """
    Plot confusion matrix (counts or normalized by true class).
    labels, preds : 1D arrays/lists of true and predicted labels (ints or str)
    class_labels : optional list of class names in index order
    normalize : if True, rows (true classes) will be normalized to sum=1
    Returns: matplotlib.Figure
    """
    cm = confusion_matrix(labels, preds)
    display = cm.astype(float)
    if normalize:
        with np.errstate(all="ignore"):
            row_sums = display.sum(axis=1, keepdims=True)
            display = np.divide(display, row_sums, where=row_sums!=0)

    n = cm.shape[0]
    if class_labels is None:
        class_labels = [str(i) for i in range(n)]

    fig, ax = plt.subplots(figsize=figsize)
    im = ax.imshow(display, interpolation='nearest', cmap=cmap)
    cbar = plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    title = "Confusion Matrix" + (" (normalized)" if normalize else " (counts)")
    ax.set_title(title)
    ax.set_xlabel("Predicted label")
    ax.set_ylabel("True label")

    ax.set_xticks(np.arange(n))
    ax.set_yticks(np.arange(n))
    ax.set_xticklabels(class_labels, rotation=45, ha="right")
    ax.set_yticklabels(class_labels)

    fmt = ".2f" if normalize else "d"
    thresh = display.max() / 2.
    for i in range(n):
        for j in range(n):
            val = display[i, j]
            txt = format(val, fmt) if not np.isnan(val) else "0"
            ax.text(j, i, txt,
                    ha="center", va="center",
                    color="white" if val > thresh else "black",
                    fontsize=9)

    plt.tight_layout()
    return fig

def _infer_history_keys(history):
    """
    Given a dict-like history, try to extract lists for:
      train_loss, val_loss, train_acc, val_acc
    Returns dict with keys possibly present.
    """
    out = {}
    # possible keys mapping
    mapping = {
        "train_loss" : ["train_loss", "loss", "training_loss"],
        "val_loss"   : ["val_loss", "validation_loss"],
        "train_acc"  : ["train_acc", "accuracy", "acc", "training_accuracy"],
        "val_acc"    : ["val_acc", "val_accuracy", "validation_accuracy"]
    }
    for target, candidates in mapping.items():
        for k in candidates:
            if isinstance(history, dict) and k in history:
                out[target] = np.array(history[k])
                break
    # if none found for train_loss but there is a top-level list (e.g., history is a tuple of two lists),
    # we don't try to guess beyond that.
    return out

def plot_train_val_curves(history=None, history_path=None, figsize=(12,5), save_path=None):
    """
    Plots training & validation loss and accuracy.
    - history: dict-like with keys as in Keras or typical PyTorch training logs.
      e.g. {'loss': [...], 'val_loss': [...], 'accuracy': [...], 'val_accuracy': [...]}
      or {'train_loss': [...], 'val_loss': [...], 'train_acc': [...], 'val_acc': [...]}
    - history_path: optional path to try loading history from (supports .pth/.pt (torch), .npy, .json)
    Returns: tuple(matplotlib.Figure for loss, matplotlib.Figure for acc or None)
    """
    hist = history

    # try to load if not provided
    if hist is None and history_path is not None:
        path = history_path
        if not os.path.exists(path):
            raise FileNotFoundError(f"history_path not found: {path}")
        ext = os.path.splitext(path)[1].lower()
        if ext in (".pth", ".pt"):
            hist = torch.load(path, map_location="cpu")
        elif ext == ".npy":
            hist = np.load(path, allow_pickle=True).item()
        elif ext == ".json":
            with open(path, "r") as f:
                hist = json.load(f)
        else:
            # try torch.load as fallback
            hist = torch.load(path, map_location="cpu")

    if hist is None:
        raise ValueError("No history provided. Pass history dict or history_path to load it.")

    # try to extract keys
    keys = _infer_history_keys(hist)
    # Build epoch array
    length = None
    for k in ["train_loss", "val_loss", "train_acc", "val_acc"]:
        if k in keys:
            length = len(keys[k])
            break
    if length is None:
        # try older keys
        # if history has 'loss' or similar
        if isinstance(hist, dict):
            possible_lengths = [len(v) for v in hist.values() if hasattr(v, "__len__")]
            length = max(possible_lengths) if possible_lengths else 0
        else:
            raise ValueError("Could not infer history lengths from provided history object.")

    epochs = np.arange(1, length+1)

    # LOSS plot
    fig_loss, ax_loss = plt.subplots(figsize=figsize)
    plotted = False
    if "train_loss" in keys:
        ax_loss.plot(epochs[:len(keys["train_loss"])], keys["train_loss"], label="train_loss")
        plotted = True
    elif "loss" in hist:
        ax_loss.plot(epochs[:len(hist["loss"])], hist["loss"], label="train_loss")
        plotted = True

    if "val_loss" in keys:
        ax_loss.plot(epochs[:len(keys["val_loss"])], keys["val_loss"], label="val_loss")
        plotted = True
    elif "val_loss" in hist:
        ax_loss.plot(epochs[:len(hist["val_loss"])], hist["val_loss"], label="val_loss")
        plotted = True

    if plotted:
        ax_loss.set_xlabel("Epoch")
        ax_loss.set_ylabel("Loss")
        ax_loss.set_title("Training & Validation Loss")
        ax_loss.legend()
        ax_loss.grid(alpha=0.2)
    else:
        plt.close(fig_loss)
        fig_loss = None

    # ACC plot (optional)
    fig_acc = None
    plotted_acc = False
    if "train_acc" in keys or "train_accuracy" in hist or "accuracy" in hist or "acc" in hist:
        fig_acc, ax_acc = plt.subplots(figsize=figsize)
        if "train_acc" in keys:
            ax_acc.plot(epochs[:len(keys["train_acc"])], keys["train_acc"], label="train_acc")
            plotted_acc = True
        elif "accuracy" in hist:
            ax_acc.plot(epochs[:len(hist["accuracy"])], hist["accuracy"], label="train_acc")
            plotted_acc = True
        elif "acc" in hist:
            ax_acc.plot(epochs[:len(hist["acc"])], hist["acc"], label="train_acc")
            plotted_acc = True

        if "val_acc" in keys:
            ax_acc.plot(epochs[:len(keys["val_acc"])], keys["val_acc"], label="val_acc")
            plotted_acc = True
        elif "val_accuracy" in hist:
            ax_acc.plot(epochs[:len(hist["val_accuracy"])], hist["val_accuracy"], label="val_acc")
            plotted_acc = True

        if plotted_acc:
            ax_acc.set_xlabel("Epoch")
            ax_acc.set_ylabel("Accuracy")
            ax_acc.set_title("Training & Validation Accuracy")
            ax_acc.legend()
            ax_acc.grid(alpha=0.2)
        else:
            plt.close(fig_acc)
            fig_acc = None

    # Save if requested
    if save_path:
        base = save_path
        if fig_loss:
            fig_loss.savefig(base + "_loss.png", bbox_inches="tight", dpi=200)
        if fig_acc:
            fig_acc.savefig(base + "_acc.png", bbox_inches="tight", dpi=200)

    return fig_loss, fig_acc

# ----------------------
# Example usage cell:
# ----------------------
# You already said you run:
# model.load_state_dict(torch.load(ckpt_dir / "best_fuzzy_emotion.pth"))
# test_acc, preds, labels = evaluate(model, test_loader)
#
# After that, run the plotting calls below. Replace CLASS_LABELS or history_path as needed.

# Example CLASS_LABELS (edit to your labels)
CLASS_LABELS = ["angry","disgust","fear","happy","sad","surprise","neutral","contempt"]

# --- CONFUSION MATRIX (counts and normalized) ---
# use your preds and labels variables
try:
    fig1 = plot_confusion_matrix(labels, preds, class_labels=CLASS_LABELS, normalize=False, figsize=(8,6))
    fig2 = plot_confusion_matrix(labels, preds, class_labels=CLASS_LABELS, normalize=True, figsize=(8,6))
    plt.show()
except Exception as e:
    print("Error plotting confusion matrix:", e)

# --- TRAIN / VAL GRAPHS ---
# Option A: pass history dict directly if you have it in memory, e.g.
# history = {'loss': [...], 'val_loss': [...], 'accuracy': [...], 'val_accuracy': [...]}
# fig_loss, fig_acc = plot_train_val_curves(history=history)

# Option B: try to load from common files:
# 'training_history.pth' (torch saved dict), 'history.npy' (dict saved with np.save), 'history.json'
history_candidates = ["training_history.pth", "history.pth", "history.pt", "history.npy", "history.json", "training_history.json"]
fig_loss = fig_acc = None
for p in history_candidates:
    if os.path.exists(p):
        try:
            fig_loss, fig_acc = plot_train_val_curves(history_path=p)
            print(f"Plotted training curves from {p}")
            break
        except Exception as e:
            print(f"Found {p} but failed to load: {e}")

if fig_loss is None and fig_acc is None:
    print("No history found or no valid curve keys detected. If you have a history dict, pass it to plot_train_val_curves(history=your_history).")
else:
    plt.show()
```
